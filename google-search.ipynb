```python
import pandas as pd
from googlesearch import search
import trafilatura
```

```python
df = pd.read_csv('/kaggle/input/fake-news/Scrapping.csv', sep=';', encoding='latin-1')
df.head()
```

```python
df['search_result'] = ''
df['extracted_content'] = ''

for index, row in df.iterrows():
    query = row['Headline']  # Gantilah 'item_column' dengan nama kolom item dalam DataFrame Anda
    
    try:
        print(f"Searching for: {query}")
        
        # Lakukan pencarian Google dan ambil hanya satu hasil pertama
        top_result = next(search(query, num_results=1), None)
#         top_result = next(search(query, num_results=1, advanced=True), None)
        
        if top_result:
            print(f"Top result: {top_result}")
            df.at[index, 'search_result'] = top_result
            content = trafilatura.extract(top_result)
            if content:
                print("Content extracted successfully.")
                df.at[index, 'extracted_content'] = content
            else:
                print("Unable to extract content.")
        else:
            print("No result found.")
        
        print("\n")  # Tambahkan baris kosong antara setiap item
    except Exception as e:
        print(f"Error searching for {query}: {e}")
```

```python
# Menyimpan DataFrame ke dalam file CSV
df.to_csv('hasil_pencarian.csv', index=False)
```
